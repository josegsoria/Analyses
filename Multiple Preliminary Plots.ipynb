```python
#Import the required modules and libraries.

import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

#This next line is important, depending on the where you run this code you may or may not have the ability to output plots. This next line shows a solution.

%matplotlib inline
```

```python
#####Plot the JES Variation#####

#####Array for the x values of the plot (same for all 3 y arrays)####

pT_true = np.array([2.5,7.5,12.5,17.5,22.5,27.5])

#####################################################################

######Arrays for the y values of the plot (3 of them)#######

array_m_10JES = np.array([0.00133084,0.0067695,0.0490047,0.0267913,0.00937606,0.00252746]) #m stands for minus

array_nominal = np.array([0.00114366,0.0033409,0.0379881,0.0339078,0.013778,0.00449162])

array_p_10JES = np.array([0.00102964,0.00228204,0.0279788,0.0360394,0.0183495,0.00714392]) #p stands for plus

############################################################

########Construct an array for the width of each pT bin########

pT_true_bin_width = [2.5]*len(array_m_10JES) #Choosing array_m_10JES amoungst existing arrays was arbitrary and similar decisions are made in different cells 

#####Construct the figure#####

fig = plt.figure(figsize=(10,8))

plt.plot(pT_true,array_m_10JES,'bo',label=r' -10% JES')
plt.errorbar(pT_true,array_m_10JES,xerr=pT_true_bin_width,fmt=None,ecolor='b')

##############################

#####Here we cheat by making a historgram by drawing vertical lines#####

plt.plot((5,5),(array_m_10JES[0],array_m_10JES[1]),'b-')
plt.plot((10,10),(array_m_10JES[1],array_m_10JES[2]),'b-')
plt.plot((15,15),(array_m_10JES[2],array_m_10JES[3]),'b-')
plt.plot((20,20),(array_m_10JES[3],array_m_10JES[4]),'b-')
plt.plot((25,25),(array_m_10JES[4],array_m_10JES[5]),'b-')

########################################################################

#####Continue to construct the figure via a similar process#####

plt.plot(pT_true,array_nominal,'go',label='Nominal')
plt.errorbar(pT_true,array_nominal,xerr=pT_true_bin_width,fmt=None,ecolor='g')

plt.plot((5,5),(array_nominal[0],array_nominal[1]),'g-')
plt.plot((10,10),(array_nominal[1],array_nominal[2]),'g-')
plt.plot((15,15),(array_nominal[2],array_nominal[3]),'g-')
plt.plot((20,20),(array_nominal[3],array_nominal[4]),'g-')
plt.plot((25,25),(array_nominal[4],array_nominal[5]),'g-')

plt.plot(pT_true,array_p_10JES,'ro',label=r'+10% JES')
plt.errorbar(pT_true,array_p_10JES,xerr=pT_true_bin_width,fmt=None,ecolor='r')

plt.plot((5,5),(array_p_10JES[0],array_p_10JES[1]),'r-')
plt.plot((10,10),(array_p_10JES[1],array_p_10JES[2]),'r-')
plt.plot((15,15),(array_p_10JES[2],array_p_10JES[3]),'r-')
plt.plot((20,20),(array_p_10JES[3],array_p_10JES[4]),'r-')
plt.plot((25,25),(array_p_10JES[4],array_p_10JES[5]),'r-')

#####Add labels to the figure#####

plt.title('JES variation by 10%',fontsize=25)
plt.xticks(fontsize=17)
plt.xlabel(r'$p^\mathrm{true}_\mathrm{T}$ (GeV/$c$)',fontsize=30)
plt.yticks(fontsize=17)
plt.legend(loc='upper right',frameon=False,prop={'size':22},numpoints=1)

##################################

fig.tight_layout()
plt.show()
fig.savefig('JES_variation.pdf')
```
